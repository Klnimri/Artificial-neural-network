```python
# Standard scientific Python imports
import numpy as np
import matplotlib.pyplot as plt

#Load data, face images and their target i.e. perosn number
data = np.load("./olivetti_faces.npy")
target = np.load("./olivetti_faces_target.npy")

#print the size of data and target
#Data has 400 images of size 64 x 64 pixels (grayscale)
#Target has information about class of all 400 images
print(data.shape)
print(target.shape)

#Identity number of unique persons (ids) in the trarget data
#For ORL / Olivetti face dataset this there are 40 people
#The dataset contains 10 images of each person. Therefore 40 x 10 = 400 images
unique_ids=np.unique(target)
print(len(unique_ids))

#Display one sample image for each person
#Display the 40 images in 4 rows with each row having 10 images
fig, axarr = plt.subplots(nrows=4, ncols=10, figsize=(18, 9))
axarr = axarr.flatten()
imgs=data
for id in unique_ids:
    img_id = id * 10
    axarr[id].imshow(imgs[img_id], cmap='gray')

#Display all 10 images of first 5 people
cols = 10
subject_ids=[0, 1, 2, 3, 4]
rows = (len(subject_ids)*10)//cols
fig, axarr = plt.subplots(nrows=rows, ncols=cols, figsize=(18, 9))
for i, subject_ids in enumerate(subject_ids):
    for j in range(cols):
        img_id = subject_ids*10 + j
        axarr[i, j].imshow(imgs[img_id], cmap='gray')

#Split the data into test and train data. Select test size = 20%
#Use sklearn library to split data into trianing and testing set
#X_train contains the image information and Y_train contains the target label (class)
#from sklearn.model_selection import train_test_split
#X_train, X_test, Y_train, Y_test = train_test_split(data, target, test_size = 0.2, random_state= 0)

#For training the data has to be split into training and validation, Split training data into training and validation set
#Set validation data size = 20% of training data
#X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size = 0.2, random_state = 0)

#Import keras for building the Convolutional Neural Network
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Dense, Dropout, Activation, Flatten

#Split the data into test and train data. Select test size = 20%
from sklearn.model_selection import train_test_split
target = np.load("./olivetti_faces_target.npy")
target = to_categorical(target, len(np.unique(target)))

#Use sklearn library to split data into trianing and testing set
#X_train contains the image information and Y_train contains the target label (class)
X_train, X_test, Y_train, Y_test = train_test_split(data, target, test_size = 0.2, random_state= 0)

#For training the data has to be split into training and validation, Split training data into training and validation set
#Set validation data size = 20% of training data
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size = 0.2, random_state = 0)

#target.shape

#Remember the third dimension is important for keras so specify that third dimension is 1
X_train = X_train.reshape(X_train.shape[0], 64, 64, 1)
X_val = X_val.reshape(X_val.shape[0], 64, 64, 1)
X_test = X_test.reshape(X_test.shape[0], 64, 64, 1)


#Define the architecture of the convolutional neural network
model = Sequential()

#Add a conv layer having 8 filters followed by a relu layer (image size 64 x 64)
model.add(Conv2D(8, (5, 5), activation='relu', input_shape=(64, 64, 1), padding='same'))
#Reduce the size of the images by 4 using maxpooling layer (image size 16 x 16)
model.add(MaxPooling2D(pool_size=(4, 4)))
#Add a conv layer having 16 filters followed by a relu layer (image size 64 x 64)
model.add(Conv2D(16, (5, 5), activation='relu', padding='same'))
#Reduce the size of the images by 4 using maxpooling layer (image size 4 x 4)
model.add(MaxPooling2D(pool_size=(4, 4)))
#Add a conv layer having 16 filters followed by a relu layer (image size 64 x 64)
model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
#Reduce the size of the images by 2 using maxpooling layer (image size 2 x 2)
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(256, activation='sigmoid'))
#Pass data through a softmax layer
model.add(Dense(40, activation='softmax'))
#Since there are more than 2 classes use categorical_crossentropy, adam optimization and optimize based upon accuracy value
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

#Print a summary of the model
model.summary()

#Train the network using the above deinfed network architecture and give accuracy results for training and validation data
H = model.fit(X_train, Y_train, validation_data=(X_val, Y_val), batch_size=32, epochs=10, verbose=1)

```

```python
from keras.layers import BatchNormalization

# Define the architecture of the convolutional neural network with Batch Normalization
model_bn = Sequential()

model_bn.add(Conv2D(8, (5, 5), activation='relu', input_shape=(64, 64, 1), padding='same'))
model_bn.add(BatchNormalization())
model_bn.add(MaxPooling2D(pool_size=(2, 2)))

model_bn.add(Conv2D(16, (5, 5), activation='relu', padding='same'))
model_bn.add(BatchNormalization())
model_bn.add(MaxPooling2D(pool_size=(2, 2)))

model_bn.add(Flatten())
model_bn.add(Dense(256, activation='sigmoid'))
model_bn.add(Dense(40, activation='softmax'))

# Compile the model
model_bn.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Print a summary of the model
model_bn.summary()

# Train the network
H_bn = model_bn.fit(X_train, Y_train, validation_data=(X_val, Y_val), batch_size=32, epochs=10, verbose=1)
```

```python
from keras.layers import Dropout

# Define the architecture of the convolutional neural network with Dropout Regularization
model_dropout = Sequential()

model_dropout.add(Conv2D(8, (5, 5), activation='relu', input_shape=(64, 64, 1), padding='same'))
model_dropout.add(MaxPooling2D(pool_size=(2, 2)))
model_dropout.add(Dropout(0.25))

model_dropout.add(Conv2D(16, (5, 5), activation='relu', padding='same'))
model_dropout.add(MaxPooling2D(pool_size=(2, 2)))
model_dropout.add(Dropout(0.25))

model_dropout.add(Flatten())
model_dropout.add(Dense(256, activation='sigmoid'))
model_dropout.add(Dense(40, activation='softmax'))

# Compile the model
model_dropout.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Print a summary of the model
model_dropout.summary()

# Train the network
H_dropout = model_dropout.fit(X_train, Y_train, validation_data=(X_val, Y_val), batch_size=32, epochs=10, verbose=1)
```

```python
from keras.optimizers import Adam
from keras.callbacks import LearningRateScheduler

# Learning rate schedule function
def lr_schedule(epoch):
    return 0.001 * (0.1 ** int(epoch / 10))

# Define the architecture of the convolutional neural network with Learning Rate Scheduler
model_lr_schedule = Sequential()

model_lr_schedule.add(Conv2D(8, (5, 5), activation='relu', input_shape=(64, 64, 1), padding='same'))
model_lr_schedule.add(MaxPooling2D(pool_size=(2, 2)))

model_lr_schedule.add(Conv2D(16, (5, 5), activation='relu', padding='same'))
model_lr_schedule.add(MaxPooling2D(pool_size=(2, 2)))

model_lr_schedule.add(Flatten())
model_lr_schedule.add(Dense(256, activation='sigmoid'))
model_lr_schedule.add(Dense(40, activation='softmax'))

# Compile the model with a learning rate schedule
model_lr_schedule.compile(loss='categorical_crossentropy', optimizer=Adam(lr=0.001), metrics=['accuracy'])

# Print a summary of the model
model_lr_schedule.summary()

# Train the network
H_lr_schedule = model_lr_schedule.fit(X_train, Y_train, validation_data=(X_val, Y_val), batch_size=32, epochs=10, verbose=1, callbacks=[LearningRateScheduler(lr_schedule)])
```

```python
import matplotlib.pyplot as plt

# Plotting function for accuracy differences
def plot_accuracy_differences(original_history, modified_history, title):
    plt.figure(figsize=(12, 8))

    original_training_diff = [acc - val_acc for acc, val_acc in zip(original_history.history['accuracy'], original_history.history['val_accuracy'])]
    original_validation_diff = [val_acc - acc for acc, val_acc in zip(original_history.history['accuracy'], original_history.history['val_accuracy'])]

    modified_training_diff = [acc - val_acc for acc, val_acc in zip(modified_history.history['accuracy'], modified_history.history['val_accuracy'])]
    modified_validation_diff = [val_acc - acc for acc, val_acc in zip(modified_history.history['accuracy'], modified_history.history['val_accuracy'])]

    plt.plot(original_training_diff, label=f'Original Training Diff')
    plt.plot(original_validation_diff, label=f'Original Validation Diff')

    plt.plot(modified_training_diff, label=f'{title} Training Diff')
    plt.plot(modified_validation_diff, label=f'{title} Validation Diff')

    plt.title(f'Accuracy Differences between Training and Validation - {title}')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy Difference')
    plt.legend()
    plt.show()

# Plot the accuracy differences for Batch Normalization
plot_accuracy_differences(H, H_bn, 'Batch Normalization')

# Plot the accuracy differences for Dropout Regularization
plot_accuracy_differences(H, H_dropout, 'Dropout Regularization')

# Plot the accuracy differences for Learning Rate Scheduler
plot_accuracy_differences(H, H_lr_schedule, 'Learning Rate Scheduler')

# Plot the accuracy differences for Increased Model Complexity
plot_accuracy_differences(H, H_complexity, 'Increased Model Complexity')

# Plot the accuracy differences for Data Augmentation
plot_accuracy_differences(H, H_data_augmentation, 'Data Augmentation')
```

```python
import matplotlib.pyplot as plt

# Function to calculate accuracy differences
def calculate_accuracy_diff(history):
    training_diff = [acc - val_acc for acc, val_acc in zip(history.history['accuracy'], history.history['val_accuracy'])]
    validation_diff = [val_acc for acc, val_acc in zip(history.history['accuracy'], history.history['val_accuracy'])]
    return training_diff, validation_diff

# Plotting function for accuracy differences
def plot_accuracy_differences(original_history, modified_history, title):
    original_training_diff, original_validation_diff = calculate_accuracy_diff(original_history)
    modified_training_diff, modified_validation_diff = calculate_accuracy_diff(modified_history)

    epochs = range(1, len(original_history.history['accuracy']) + 1)

    plt.figure(figsize=(12, 8))

    plt.plot(epochs, original_training_diff, label=f'Original Training Diff')
    plt.plot(epochs, original_validation_diff, label=f'Original Validation Diff')

    plt.plot(epochs, modified_training_diff, label=f'{title} Training Diff')
    plt.plot(epochs, modified_validation_diff, label=f'{title} Validation Diff')

    plt.title(f'Accuracy Differences between Training and Validation - {title}')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy Difference')
    plt.legend()
    plt.show()

# Plot the accuracy differences for ELU Activation
plot_accuracy_differences(H, H_elu, 'ELU Activation')
```
